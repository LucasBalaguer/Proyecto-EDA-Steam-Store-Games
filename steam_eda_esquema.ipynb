## 📌 Introducción

Steam es la plataforma de distribución de videojuegos para PC más grande del mundo, con más de 50.000 juegos disponibles y millones de usuarios activos.

Este proyecto realiza un Análisis Exploratorio de Datos (EDA) sobre el catálogo de Steam con el objetivo de entender qué factores influyen en el éxito de un videojuego: su valoración, su popularidad y su rendimiento comercial.

El análisis parte de preguntas de negocio concretas definidas antes de explorar los datos, siguiendo una metodología profesional orientada a la toma de decisiones.

---

## 🎯 Objetivos del Proyecto

- Identificar tendencias en el mercado de videojuegos de PC.
- Analizar la relación entre precio, valoraciones y popularidad.
- Detectar qué géneros y desarrolladores dominan el mercado.
- Practicar el flujo completo de un EDA: carga, limpieza, análisis y visualización.
- Generar visualizaciones claras y atractivas para comunicar los hallazgos.

---

## ❓ Preguntas de Negocio

Las siguientes preguntas guían el análisis. Fueron definidas **antes** de explorar los datos:

1. ¿Cómo ha evolucionado la cantidad de juegos publicados por año?
2. ¿Qué géneros dominan el mercado y cuáles tienen mejor valoración?
3. ¿Existe relación entre el precio de un juego y su valoración?
4. ¿Qué desarrolladores tienen mejor ratio calidad-precio?
5. ¿Los juegos gratuitos tienen peor valoración que los de pago?

> 💡 Durante la exploración pueden surgir nuevas preguntas que se irán añadiendo aquí.

## 📋 Diccionario de Columnas

### Dataset principal: `steam.csv`

| Columna | Tipo | Descripción |
|---|---|---|
| `appid` | int | Identificador único del juego en Steam |
| `name` | str | Nombre del juego |
| `release_date` | datetime | Fecha de lanzamiento |
| `english` | int | 1 si está disponible en inglés, 0 si no |
| `developer` | str | Nombre del estudio desarrollador |
| `publisher` | str | Nombre del publisher (distribuidor) |
| `platforms` | str | Plataformas disponibles (windows, mac, linux) |
| `required_age` | int | Edad mínima requerida (0 si no hay restricción) |
| `categories` | str | Categorías de Steam (multijugador, co-op, etc.) separadas por `;` |
| `genres` | str | Géneros del juego separados por `;` |
| `steamspy_tags` | str | Tags asignados por los usuarios en SteamSpy, separados por `;` |
| `achievements` | int | Número de logros disponibles en el juego |
| `positive_ratings` | int | Número total de valoraciones positivas |
| `negative_ratings` | int | Número total de valoraciones negativas |
| `average_playtime` | int | Tiempo medio de juego en minutos |
| `median_playtime` | int | Tiempo mediano de juego en minutos |
| `owners` | str | Rango estimado de propietarios del juego (ej: '10000000-20000000') |
| `price` | float | Precio en dólares |

### Columnas creadas durante la limpieza

| Columna | Tipo | Descripción |
|---|---|---|
| `release_year` | int | Año de lanzamiento extraído de `release_date` |
| `owners_mean` | float | Valor medio del rango de `owners`, para uso numérico |
| `positive_ratio` | float | Porcentaje de valoraciones positivas sobre el total |

## 📝 Notas y Limitaciones

- Los datos fueron extraídos en **mayo de 2019**, por lo que no incluyen juegos lanzados después de esa fecha.
- La columna `owners` es una **estimación** de SteamSpy, no un dato oficial de Valve.
- Algunos juegos pueden tener pocos ratings, lo que hace que su `positive_ratio` no sea representativo.
